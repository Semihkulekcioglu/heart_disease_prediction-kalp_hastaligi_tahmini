## 5. Hedef DeÄŸiÅŸken DaÄŸÄ±lÄ±mÄ±


```python
# Hedef deÄŸiÅŸken daÄŸÄ±lÄ±mÄ±
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Count plot
target_counts = df['target'].value_counts()
axes[0].bar(['SaÄŸlÄ±klÄ± (0)', 'Hasta (1)'], target_counts.values, color=['green', 'red'], alpha=0.7)
axes[0].set_title('Hedef DeÄŸiÅŸken DaÄŸÄ±lÄ±mÄ±', fontsize=14, fontweight='bold')
axes[0].set_ylabel('SayÄ±', fontsize=12)
axes[0].grid(axis='y', alpha=0.3)

# DeÄŸerleri yazdÄ±r
for i, v in enumerate(target_counts.values):
    axes[0].text(i, v + 5, str(v), ha='center', fontweight='bold', fontsize=12)

# Pie chart
axes[1].pie(target_counts.values, labels=['SaÄŸlÄ±klÄ± (0)', 'Hasta (1)'], 
            autopct='%1.1f%%', startangle=90, colors=['green', 'red'], explode=(0.05, 0.05))
axes[1].set_title('Hedef DeÄŸiÅŸken OranÄ±', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()

print(f"\nðŸ“Š Hedef DeÄŸiÅŸken Ä°statistikleri:")
print(f"  SaÄŸlÄ±klÄ± (0): {target_counts[0]} ({target_counts[0]/len(df)*100:.2f}%)")
print(f"  Hasta (1): {target_counts[1]} ({target_counts[1]/len(df)*100:.2f}%)")
print(f"  Toplam: {len(df)}")
```

## 6. Ã–zellik DaÄŸÄ±lÄ±mlarÄ±


```python
# SÃ¼rekli deÄŸiÅŸkenlerin daÄŸÄ±lÄ±mÄ±
continuous_features = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']

fig, axes = plt.subplots(2, 3, figsize=(16, 10))
axes = axes.ravel()

for idx, feature in enumerate(continuous_features):
    axes[idx].hist(df[feature], bins=30, color='skyblue', edgecolor='black', alpha=0.7)
    axes[idx].set_title(f'{feature.upper()} DaÄŸÄ±lÄ±mÄ±', fontsize=12, fontweight='bold')
    axes[idx].set_xlabel(feature, fontsize=10)
    axes[idx].set_ylabel('Frekans', fontsize=10)
    axes[idx].grid(axis='y', alpha=0.3)
    
    # Ä°statistikler ekle
    mean_val = df[feature].mean()
    median_val = df[feature].median()
    axes[idx].axvline(mean_val, color='red', linestyle='--', linewidth=2, label=f'Ortalama: {mean_val:.1f}')
    axes[idx].axvline(median_val, color='green', linestyle='--', linewidth=2, label=f'Medyan: {median_val:.1f}')
    axes[idx].legend()

# Son subplot'u gizle
axes[5].axis('off')

plt.tight_layout()
plt.show()
```

```python
# Kategorik deÄŸiÅŸkenlerin daÄŸÄ±lÄ±mÄ±
categorical_features = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope']

fig, axes = plt.subplots(2, 3, figsize=(16, 10))
axes = axes.ravel()

for idx, feature in enumerate(categorical_features):
    value_counts = df[feature].value_counts().sort_index()
    axes[idx].bar(value_counts.index, value_counts.values, color='coral', alpha=0.7, edgecolor='black')
    axes[idx].set_title(f'{feature.upper()} DaÄŸÄ±lÄ±mÄ±', fontsize=12, fontweight='bold')
    axes[idx].set_xlabel(feature, fontsize=10)
    axes[idx].set_ylabel('SayÄ±', fontsize=10)
    axes[idx].grid(axis='y', alpha=0.3)
    
    # DeÄŸerleri yazdÄ±r
    for i, v in zip(value_counts.index, value_counts.values):
        axes[idx].text(i, v + 1, str(v), ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

## 7. Korelasyon Analizi


```python
# Korelasyon matrisi
plt.figure(figsize=(14, 12))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', 
            square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Ã–zellikler ArasÄ± Korelasyon Matrisi', fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()

# Hedef deÄŸiÅŸken ile en yÃ¼ksek korelasyona sahip Ã¶zellikler
print("\nðŸŽ¯ Hedef DeÄŸiÅŸken ile Korelasyon:")
target_corr = correlation_matrix['target'].sort_values(ascending=False)
print(target_corr)
```

```python
# En Ã¶nemli korelasyonlarÄ± gÃ¶rselleÅŸtir
top_features = target_corr[1:6].index  # Ä°lk 5 Ã¶zellik (target hariÃ§)

plt.figure(figsize=(12, 6))
bars = plt.barh(range(len(top_features)), target_corr[top_features].values, color='steelblue', alpha=0.7)
plt.yticks(range(len(top_features)), top_features)
plt.xlabel('Korelasyon DeÄŸeri', fontsize=12)
plt.title('Target ile En YÃ¼ksek Korelasyona Sahip 5 Ã–zellik', fontsize=14, fontweight='bold')
plt.grid(axis='x', alpha=0.3)

# DeÄŸerleri yazdÄ±r
for i, v in enumerate(target_corr[top_features].values):
    plt.text(v + 0.01, i, f'{v:.3f}', va='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

## 8. Ã–zellik-Hedef Ä°liÅŸkileri


```python
# YaÅŸ ve kalp hastalÄ±ÄŸÄ± iliÅŸkisi
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
for target_val in [0, 1]:
    data = df[df['target'] == target_val]['age']
    plt.hist(data, bins=20, alpha=0.6, label=f'Target = {target_val}', edgecolor='black')
plt.xlabel('YaÅŸ', fontsize=12)
plt.ylabel('Frekans', fontsize=12)
plt.title('YaÅŸ DaÄŸÄ±lÄ±mÄ± - Hedef DeÄŸiÅŸkene GÃ¶re', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(alpha=0.3)

plt.subplot(1, 2, 2)
df.boxplot(column='age', by='target', ax=plt.gca())
plt.xlabel('Hedef DeÄŸiÅŸken', fontsize=12)
plt.ylabel('YaÅŸ', fontsize=12)
plt.title('YaÅŸ - Hedef DeÄŸiÅŸken Box Plot', fontsize=14, fontweight='bold')
plt.suptitle('')

plt.tight_layout()
plt.show()
```

```python
# Cinsiyet ve kalp hastalÄ±ÄŸÄ± iliÅŸkisi
cross_tab = pd.crosstab(df['sex'], df['target'], normalize='index') * 100

plt.figure(figsize=(10, 6))
cross_tab.plot(kind='bar', color=['green', 'red'], alpha=0.7, edgecolor='black')
plt.xlabel('Cinsiyet (0=KadÄ±n, 1=Erkek)', fontsize=12)
plt.ylabel('Oran (%)', fontsize=12)
plt.title('Cinsiyet - Kalp HastalÄ±ÄŸÄ± Ä°liÅŸkisi', fontsize=14, fontweight='bold')
plt.legend(['SaÄŸlÄ±klÄ±', 'Hasta'], loc='upper right')
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print("\nðŸ“Š Cinsiyet - Hedef Ã‡apraz Tablo:")
print(pd.crosstab(df['sex'], df['target'], margins=True))
```

```python
# GÃ¶ÄŸÃ¼s aÄŸrÄ±sÄ± tipi ve kalp hastalÄ±ÄŸÄ± iliÅŸkisi
cross_tab_cp = pd.crosstab(df['cp'], df['target'], normalize='index') * 100

plt.figure(figsize=(10, 6))
cross_tab_cp.plot(kind='bar', color=['green', 'red'], alpha=0.7, edgecolor='black')
plt.xlabel('GÃ¶ÄŸÃ¼s AÄŸrÄ±sÄ± Tipi (cp)', fontsize=12)
plt.ylabel('Oran (%)', fontsize=12)
plt.title('GÃ¶ÄŸÃ¼s AÄŸrÄ±sÄ± Tipi - Kalp HastalÄ±ÄŸÄ± Ä°liÅŸkisi', fontsize=14, fontweight='bold')
plt.legend(['SaÄŸlÄ±klÄ±', 'Hasta'], loc='upper right')
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print("\nðŸ“Š GÃ¶ÄŸÃ¼s AÄŸrÄ±sÄ± - Hedef Ã‡apraz Tablo:")
print(pd.crosstab(df['cp'], df['target'], margins=True))
```

```python
# Maksimum kalp atÄ±ÅŸ hÄ±zÄ± ve hedef iliÅŸkisi
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
for target_val in [0, 1]:
    data = df[df['target'] == target_val]['thalach']
    plt.hist(data, bins=20, alpha=0.6, label=f'Target = {target_val}', edgecolor='black')
plt.xlabel('Maksimum Kalp AtÄ±ÅŸ HÄ±zÄ±', fontsize=12)
plt.ylabel('Frekans', fontsize=12)
plt.title('Maksimum Kalp AtÄ±ÅŸ HÄ±zÄ± - Hedef DeÄŸiÅŸkene GÃ¶re', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(alpha=0.3)

plt.subplot(1, 2, 2)
df.boxplot(column='thalach', by='target', ax=plt.gca())
plt.xlabel('Hedef DeÄŸiÅŸken', fontsize=12)
plt.ylabel('Maksimum Kalp AtÄ±ÅŸ HÄ±zÄ±', fontsize=12)
plt.title('Maksimum Kalp AtÄ±ÅŸ HÄ±zÄ± Box Plot', fontsize=14, fontweight='bold')
plt.suptitle('')

plt.tight_layout()
plt.show()
```
